5번

```python
df['year_for_graph'] = pd.DatetimeIndex(df['일시']).year

yearly_temp = df.groupby('year_for_graph')['평균 기온'].mean()

data_for_graph = pd.DataFrame(yearly_temp)
print(data_for_graph)
data_for_graph.plot(xlabel = '년도', ylabel = '일별 평균 기온의 년도별 평균 기온', color = 'skyblue')
```

8번

```python
df['year_for_graph'] = pd.DatetimeIndex(df['일시']).year
average_temp = df.groupby('year_for_graph')['평균 기온'].mean()
print(average_temp)
print()

selected_data = average_temp.loc[2015:2016]
hottest_year = selected_data.idxmax() 

print(selected_data)
print()
print("2015년과 2016년 중 더 더웠던 해는: ", hottest_year)
```

9번

```python
df['month_for_graph'] = pd.DatetimeIndex(df['일시']).month
month_average_temp = df.groupby('month_for_graph')['평균 기온'].sum()

max_temp = month_average_temp.max()
min_temp = month_average_temp.min()

max_t = month_average_temp[month_average_temp == max_temp].index[0]  # 최고 기온을 기록한 달을 찾습니다.
min_t = month_average_temp[month_average_temp == min_temp].index[0]  # 최저 기온을 기록한 달을 찾습니다.

print(month_average_temp)
print()
print("최고 기온을 기록한 달:", max_t, "월")
print("최저 기온을 기록한 달:", min_t, "월")
```
